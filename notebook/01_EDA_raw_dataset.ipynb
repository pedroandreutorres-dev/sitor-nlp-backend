### Ejecución de la Limpieza y Transformación de Datos (`df_01`)**

Una vez finalizada la fase de auditoría y análisis exploratorio, se ejecuta el pipeline de limpieza sobre el primer dataframe (`df_01`) para estandarizar su estructura de cara a la futura unificación.

**Acciones que se ejecutan en la siguiente celda de código:**

1. **Selección de Columnas Core:** Nos quedamos únicamente con **7 columnas** (`body`, `subject`, `answer`, `queue`, `type`, `priority`, `language`), eliminando por completo las columnas de `tag_1` a `tag_8` y la columna `version` por considerarse ruido o redundancia algorítmica.
2. **Filtrado de Dominio (Filas):** Eliminamos los registros de las colas ajenas a nuestro caso de negocio de soporte tecnológico y facturación (`Human Resources`, `Returns and Exchanges` y `General Inquiry`).
3. **Tratamiento de Valores Nulos:**
   * Imputación en `subject`: Los valores faltantes se rellenan con la cadena `"No Subject"`.
   * Depuración en `answer`: Se eliminan las filas con nulos en esta columna para proteger el futuro módulo RAG.
4. **Saneamiento del Índice:** Aplicación de `.reset_index(drop=True)` para reordenar los índices tras el descarte de registros.

```python
# Eliminamos la columna 'version' del df_01
df_01 = df_01.drop(columns=['version'])

# Eliminamos todas las columnas tag del df_01
df_01 = df_01.drop(columns=[col for col in df_01.columns if col.startswith('tag_')])

# De la columna queue del df_01 eliminamos las filas donde aparezca Human Resources, Returns and Exchanges y General Inquiry
df_01 = df_01[~df_01['queue'].isin(['Human Resources', 'Returns and Exchanges', 'General Inquiry'])]

# En la columna 'subject' del df_01 rellenamos los valores nulos con el valor 'No Subject'
df_01['subject'] = df_01['subject'].fillna('No Subject')

# Eliminamos las filas del df_01 donde haya nulos en la columna 'answer'
df_01 = df_01.dropna(subset=['answer'])

# Reseteamos el indice del df_01 después de eliminar filas
df_01 = df_01.reset_index(drop=True)

# Mostramos el número total de filas y columnas actuales
print(f"\nNúmero total de filas y columnas actuales del df_01: {df_01.shape}")

# Mostramos el conteo de valores nulos
print("\nConteo de valores nulos despues de la limpieza del df_01:")
print(df_01.isnull().sum())

# Mostramos la distribución de frecuencias actualizada de la columna queue
print("\nDistribución de frecuencias de la columna 'queue' después de la limpieza:")
print(df_01['queue'].value_counts())

# Mostramos el primer y último índice del df_01 después de la limpieza
print(f"\nPrimer índice del df_01 después de la limpieza: {df_01.index}")


```

### Ejecución de la Limpieza y Transformación de Datos (`df_02`)

Procedemos a ejecutar el pipeline de limpieza sobre el segundo dataframe (`df_02`) para alinear su estructura con `df_01`. Esto asegura que la futura concatenación se realice sobre un esquema validado y coherente con el dominio de negocio definido para el MVP[cite: 3].

**Acciones a ejecutar en la siguiente celda:**

1. **Purga de ruido categórico:** Eliminación de todas las columnas de etiquetas (`tag_1` a `tag_8`).
2. **Filtrado de Dominio:** Exclusión de los registros pertenecientes a colas fuera del alcance del proyecto (`Human Resources`, `Returns and Exchanges`, `General Inquiry`)[cite: 3].
3. **Tratamiento de Valores Nulos:**
   * Imputación del valor estático `"No Subject"` en la columna `subject`.
   * Eliminación de las filas que contengan valores nulos en las columnas `body` (imprescindible para la vectorización) y `answer` (necesaria para el módulo RAG opcional)[cite: 3].
4. **Reconstrucción Estructural:** Reseteo del índice eliminando la numeración anterior para evitar columnas residuales.

```python
# Elimina cualquier columna cuyo nombre empiece por la cadena de texto tag_.
df_02 = df_02.drop(columns=[col for col in df_02.columns if col.startswith('tag_')])

# Nos quedamos únicamente con las filas cuya columna queue no contenga las tres categorías descartadas (`Human Resources`, `Returns and Exchanges`, `General Inquiry`)
df_02 = df_02[~df_02['queue'].isin(['Human Resources', 'Returns and Exchanges', 'General Inquiry'])]

# Rellenamos los nulos de la columna subject con el valor 'No Subject'.
df_02['subject'] = df_02['subject'].fillna('No Subject')

# Eliminamos cualquier fila que tenga un nulo en la columna answer o en la columna body.
df_02 = df_02.dropna(subset=['answer', 'body'])

# Resetamos el índice del dataframe después de eliminar filas.
df_02 = df_02.reset_index(drop=True)

# Las dimensiones finales: Necesito ver la tupla con el número resultante de filas y el número exacto de columnas.
print(f"Dimensiones finales del df_02: {df_02.shape}")

# COmprobamos los nulos de cada columna del df_02
print("\nConteo de valores nulos del df_02:")   
print(df_02.isnull().sum())

# Comprobamos los valores unicos de la columna queue del df_02
print("\nValores únicos de la columna 'queue' del df_02 después de la limpieza:")
print(df_02['queue'].unique())
```
