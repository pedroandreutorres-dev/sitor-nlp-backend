# Carga y union de los diferentes datasets

```python
import os
import pandas as pd

#cargamos los csv que están en la carpeta data/raw
data_files = [f for f in os.listdir("../data/raw") if f.endswith(".csv")]

# #voy a listar el contenido de data:files
# for file in data_files:
#     print(file)

# Creo un df_x para cada csv que se encuentra en data_files
for file in data_files:
    df_name = "df_" + file.split(".")[0]
    globals()[df_name] = pd.read_csv("../data/raw/" + file)

# Muestro las columnas y tipo de datos de cada una en cada dataframe
for file in data_files:
    df_name = "df_" + file.split(".")[0]
    print(f"Dataframe: {df_name}")
    print(globals()[df_name].info())
    print("###########################################################")
    print("\n")

```
